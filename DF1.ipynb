## The .astype() Method

```python
## cannot have any nulls in the df
```

```python
nba = pd.read_csv('//Users//B//Documents//GIT//Data-analysis-with-Pandas//CODES//pandas//nba.csv').dropna(how = 'all')
#dropped all null values at end
#astype must not conmtain nulls to execute
```

```python
nba.isnull().sum()
```

```python
nba.isnull().sum().sum()
```

```python
nba['Salary'].fillna(77777, inplace = True)
nba['College'].fillna('HS', inplace= True)
nba.head(3)
```

```python
#find out about data types
nba.dtypes
#objects == strings
#pandas must import ints as floats, however floats are larger size
```

```python
nba.info()
#also mentions if we have nulls
```

```python
nba['Salary'] = nba['Salary'].astype("int")
#to be safe wrap the data type in duble quotes
```

```python
#replace data types
nba['Age'] = nba['Age'].astype("int")
nba['Number'] = nba['Number'].astype("int")
nba['Weight'] = nba['Weight'].astype("int")
nba.head()
```

```python
nba.info()
```

```python
##categories, because there are not many unique, memory can be saved by transforming them
nba['Position'].unique()
```

```python
nba['Position'] = nba['Position'].astype("category")
nba["Team"] = nba['Team'].astype("category")
```

```python
nba.info()
```

```python
nba['Team'].unique()
```
